```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.random(6000)*100 #losowe wartości [0-1] znormalizowane do 100
y = np.random.random(6000)*100

#scatter plot:
plt.scatter(x,y, c='violet', marker='.', s=25, alpha=0.3) #alpha - uzywana, gdy dane nakładają się na siebie to są intensywnego koloru, a o mnijeszej gęstości ulokowania bardziej transparentne
```

```python
#classic line plot:
years=[2000 + x for x in range(24)] #lista lat genrowanych autoamtycznie od 2000 do 2023
weight=[]
for x in range(len(years)):
    weight.append(np.random.randint(40,120))

plt.plot(years, weight, c='orange', lw='2', linestyle='--')
```

```python
#plot customization:
from matplotlib import style
year=[2010+x for x in range(14)]
progress=np.random.randint(10, 80, len(year))
progress2=np.random.randint(10, 80, len(year))
progress=sorted(progress)
progress2=sorted(progress2)
plt.plot(year, progress, label='Martha')
plt.title('Progession of IMAGINE wallet', fontsize=16, fontname='Algerian')
plt.xlabel('years')
plt.ylabel('US dollars')
y_presentation=list(range(10,80,4))
plt.yticks(y_presentation, [f"${x}k " for x in y_presentation])
plt.plot(year, progress2, label='Eric')
plt.legend(loc='lower right')
style.use('dark_background')
plt.savefig("C:\\Users\\ASUS\\Desktop\\DANE 03\\segregator\\programowanie\\pythonProject\\data science Python\\plot.jpg", dpi=300)
plt.show()
```

```python
#bar plot:
style.use('default')
x_data=['C++', 'Python', 'Ruby', 'C#', 'JavaScript', 'Java', 'Matlab', 'PHP', 'Swift', 'Kotlin']
y_data=[]
for y in range(len(x_data)):
    y_data.append(np.random.randint(0,100))

plt.bar(x_data, y_data, width=0.6, align='center', color='red', edgecolor='black', lw=2)
plt.tight_layout() #aby etykiety na osiach sie nie nakładały na siebie
```

```python
#histogram:
ages=np.random.normal(20, 2, 1000) #generowanie próbek z r. normalnego parametry:(średnia wart. oczekiwana, odch. stand., ilość próbek)
plt.hist(ages, bins=20) #bins - ilość słupków w histogramie
plt.show()
plt.hist(ages, bins=[ages.min(), 18, 20, ages.max()])
plt.show()
plt.hist(ages, bins=20, cumulative=True) #oznacza: na osi X wiek, a na Y ile osób jest poniżej danej wartości z X
plt.show()
```

```python
#pie chart:
plt.pie(y_data, explode=[0,0.2,0,0,0,0,0,0,0,0], autopct='%.2f%%', pctdistance=0.75,startangle=90, labels=None) #explode - wyróżnia dany kawałek
plt.legend(labels=x_data, prop={'size': 5})
plt.show()
```

```python
#boxplot:
heights=np.random.normal(175, 15, 1000)
plt.boxplot(heights)
plt.show()
```

```python
#multiple plots:
x1, y1=np.random.random(100), np.random.random(100)
x2, y2=np.arange(100), np.random.random(100)
x3, y3=np.random.random(100), np.random.random(100)
plt.figure(1)
plt.scatter(x1, y1)
plt.figure(2)
plt.plot(x2,y2)
plt.figure(3)
plt.scatter(x3,y3)
plt.show()
#subplot:
fig, axs = plt.subplots(3,1)
axs[0].scatter(x1,y1, s=0.1)
axs[1].plot(x2,y2)
axs[2].scatter(x3,y3, s=0.3)
fig.suptitle('Wykresy trzy')
plt.show()
```

```python
#3d plots:
style.use('default')
axis=plt.axes(projection='3d')

xx=np.random.random(100)
yy=np.random.random(100)
zz=np.random.random(100)
axis.scatter(xx,yy,zz)
axis.set_title('3D wykres')
axis.set_xlabel('x')
axis.set_ylabel('y')
axis.set_zlabel('z')
plt.tight_layout()
plt.show()

axis2=plt.axes(projection='3d')
xx1=np.arange(0,100, 0.2)
yy1=np.sin(xx1)
zz1=np.cos(xx1)
axis2.plot(xx1, yy1, zz1)
plt.show()

i,j=np.arange(-10,10, 0.1), np.arange(-10,10, 0.1)
axis3=plt.axes(projection='3d')
a,b = np.meshgrid(i,j)
c=np.sin(a)*np.cos(b)
axis3.plot_surface(a,b,c, cmap='Spectral')
plt.show()
```

```python
#animated plots:
import random
from IPython.display import clear_output #tylko do wersji jupyer notebook
head_tails=[0,0]
plt.figure()
for _ in range(500):
    head_tails[random.randint(0,1)]+=1
    plt.bar(["Heads", "Tails"], head_tails, color=["blue", "violet"])
    plt.ylim(0,500)
    plt.pause(0.01)
    clear_output(wait=True) #aby nie kumulować 500 wykresów w outpucie, każdy poprzedni jest usuwany
plt.show()
```
