```python
import pandas as pd

dict1 ={'Name':['Piotr','Michalina','Katarzyna','Zenon','Kacper',
                'Hanna','Sylwia'],
        'Marks':[98,89,99,87,90,83,99],
        'Gender':['Male','Female','Female','Male','Male',
                'Female','Female']
               }
df1=pd.DataFrame(dict1)
df1
```

```python
# 7. Create a bar graph showing average ratings by gender
import matplotlib.pyplot as plt
male_marks=df1[df1['Gender']=='Male']['Marks'].mean()
female_marks=df1[df1['Gender']=='Female']['Marks'].mean()
print(male_marks)
print(female_marks)
plt.ylim(80,100)
plt.bar(['Female', 'Male'], [female_marks, male_marks], color=['red', 'blue'], width=0.5)

# plt.bar()
```

```python
# 8. Create a grade histogram to see the distribution of grades among students
plt.hist(df1['Marks'], width=1, align='mid')
plt.legend([sorted(list(df1['Marks'].unique()))])
plt.xlabel('Grades')
plt.ylabel('Number of students')
plt.title('Histogram')
plt.show()
plt.hist(df1['Marks'], cumulative=True)
plt.title('Cumulative histogram')
plt.xlabel('Grades')
plt.ylabel('Number of students')
```
