```python
#matlplotlib:
import matplotlib.pyplot as plt 
x=list(range(0,10))
y=list(range(-10,0))

plt.plot(x,y) #wygenerowanie klasycznego wykresu funkcji dla parametrów x i y
```

```python
a=[-17, -5, 12, 33, -20, -2]
b=[-5, 3, 1, 7, 1, 9]
plt.plot(a,b) #tą samą metodą nie można wyświetlić wykrsu, będzie nieprawidłowy
plt.axis([-4, 35, 0, 7.5]) #skalowanie wykresu (wycięcie fragmentu plotu o podanych koordynatach)
plt.plot(a,b)
```

```python
#opis wykresu
plt.title('triangle')
plt.xlabel('parameter X')
plt.ylabel('parameter Y')
plt.axis([-4, 35, 0, 7.5])
plt.xticks((0,5,10,15,20,25,30,35), ('p', 'y', 't','h','o','n','i','s' )) #zamienia wartości z osi x na podane
plt.plot(a,b, color='red')
```

```python
#inne wykresy:
plt.hist(a)
plt.hist(b) # dodanie drugiego histogramu z innym parametrem doda wyniki do tego poprzedniego
```
