```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
variable=pd.DataFrame({'normal': np.random.normal(size=100),
                       'gamma': np.random.gamma(1, size=100),
                       'poisson': np.random.poisson(size=100)
                       })
variable.cumsum(0).plot() #suma skumulowana - z każdą iteracją przez wiersze danej kolumny (0) dodaje wartości do sumy z poprzedniej iteracji
```

```python
variable.cumsum(0).plot(subplots=True)
```
